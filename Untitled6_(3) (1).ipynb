```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader

# ==========================================
# TASK-0: DATASET PREPARATION
# ==========================================
class NameDataset(Dataset):
    def __init__(self, file_path):
        # The PDF requires 1000 Indian names [cite: 50]
        try:
            with open(file_path, 'r') as f:
                self.names = [line.strip() for line in f.readlines() if line.strip()]
        except FileNotFoundError:
            # Fallback for code execution if file is missing
            self.names = ["^arjun$", "^advait$", "^vihaan$", "^sanya$", "^diya$"] * 200

        self.chars = sorted(list(set(''.join(self.names))))
        self.char_to_int = {c: i for i, c in enumerate(self.chars)}
        self.int_to_char = {i: c for i, c in enumerate(self.chars)}
        self.vocab_size = len(self.chars)

    def __len__(self):
        return len(self.names)

    def __getitem__(self, idx):
        name = self.names[idx]
        encoded = [self.char_to_int[char] for char in name]
        # Shifted sequences for autoregressive learning (Input: name[0:-1], Target: name[1:])
        return torch.tensor(encoded[:-1]), torch.tensor(encoded[1:])

def collate_fn(batch):
    xs, ys = zip(*batch)
    xs_padded = nn.utils.rnn.pad_sequence(xs, batch_first=True, padding_value=0)
    ys_padded = nn.utils.rnn.pad_sequence(ys, batch_first=True, padding_value=-1)
    return xs_padded, ys_padded

# ==========================================
# TASK-1: MODEL IMPLEMENTATIONS (FROM SCRATCH)
# ==========================================

# --- [1] Multi-Layer Vanilla RNN ---
# Architecture: Stacks 2 RNN layers where layer 2 uses layer 1's hidden states as input.
class VanillaRNNScratch(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size, num_layers=2):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.embedding = nn.Embedding(vocab_size, embed_size)

        # Parameter storage for multiple layers [cite: 57]
        self.layers = nn.ModuleList()
        for i in range(num_layers):
            input_dim = embed_size if i == 0 else hidden_size
            layer_params = nn.ParameterDict({
                'W_ih': nn.Parameter(torch.randn(input_dim, hidden_size) * 0.01),
                'W_hh': nn.Parameter(torch.randn(hidden_size, hidden_size) * 0.01),
                'b_h': nn.Parameter(torch.zeros(hidden_size))
            })
            self.layers.append(layer_params)

        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x, return_hidden=False):
        batch_size, seq_len = x.shape
        current_input = self.embedding(x)

        # Iterate through layers
        for l in range(self.num_layers):
            h_t = torch.zeros(batch_size, self.hidden_size).to(x.device)
            params = self.layers[l]
            layer_outputs = []

            # Recurrence loop across time steps
            for t in range(seq_len):
                x_t = current_input[:, t, :]
                # Manual RNN formula: tanh(x*W + h*W + b)
                h_t = torch.tanh(x_t @ params['W_ih'] + h_t @ params['W_hh'] + params['b_h'])
                layer_outputs.append(h_t.unsqueeze(1))

            current_input = torch.cat(layer_outputs, dim=1) # Hidden states become input for next layer

        logits = self.fc(current_input)
        if return_hidden:
            return logits, current_input # Return hidden states for Attention mechanism
        return logits, None

# --- [2] Multi-Layer Bidirectional LSTM ---
# Architecture: Concatenates forward and backward pass hidden states at each layer[cite: 54].
class BLSTMScratch(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size, num_layers=2):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.embedding = nn.Embedding(vocab_size, embed_size)

        self.fwd_params = nn.ModuleList()
        self.bwd_params = nn.ModuleList()

        for i in range(num_layers):
            input_dim = embed_size if i == 0 else hidden_size * 2
            self.fwd_params.append(self._make_lstm_params(input_dim))
            self.bwd_params.append(self._make_lstm_params(input_dim))

        self.fc = nn.Linear(hidden_size * 2, vocab_size)

    def _make_lstm_params(self, input_dim):
        return nn.ParameterDict({
            'W': nn.Parameter(torch.randn(input_dim + self.hidden_size, 4 * self.hidden_size) * 0.01),
            'b': nn.Parameter(torch.zeros(4 * self.hidden_size))
        })

    def _lstm_cell(self, x_t, h_prev, c_prev, params):
        combined = torch.cat((x_t, h_prev), dim=1)
        gates = combined @ params['W'] + params['b']
        i, f, g, o = torch.split(gates, self.hidden_size, dim=1)
        # LSTM Gate Logic
        c_next = torch.sigmoid(f) * c_prev + torch.sigmoid(i) * torch.tanh(g)
        h_next = torch.sigmoid(o) * torch.tanh(c_next)
        return h_next, c_next

    def forward(self, x):
        batch_size, seq_len = x.shape
        current_input = self.embedding(x)

        for l in range(self.num_layers):
            f_p, b_p = self.fwd_params[l], self.bwd_params[l]
            # Forward
            h_f, c_f = torch.zeros(batch_size, self.hidden_size).to(x.device), torch.zeros(batch_size, self.hidden_size).to(x.device)
            f_outs = []
            for t in range(seq_len):
                h_f, c_f = self._lstm_cell(current_input[:, t, :], h_f, c_f, f_p)
                f_outs.append(h_f.unsqueeze(1))
            # Backward
            h_b, c_b = torch.zeros(batch_size, self.hidden_size).to(x.device), torch.zeros(batch_size, self.hidden_size).to(x.device)
            b_outs = []
            for t in reversed(range(seq_len)):
                h_b, c_b = self._lstm_cell(current_input[:, t, :], h_b, c_b, b_p)
                b_outs.insert(0, h_b.unsqueeze(1))

            current_input = torch.cat([torch.cat(f_outs, dim=1), torch.cat(b_outs, dim=1)], dim=2)

        return self.fc(current_input), None

# --- [3] RNN + Basic Attention ---
# Architecture: 2-Layer RNN followed by a dot-product attention mechanism[cite: 55].
class AttentionRNNScratch(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size):
        super().__init__()
        self.rnn = VanillaRNNScratch(vocab_size, embed_size, hidden_size, num_layers=2)
        self.W_attn = nn.Linear(hidden_size, hidden_size)
        self.fc = nn.Linear(hidden_size * 2, vocab_size)

    def forward(self, x):
        # Fix: Get hidden states (current_input) instead of just logits
        _, h_stack = self.rnn(x, return_hidden=True)

        # Self-Attention logic: Score = Query * Keys^T
        query = self.W_attn(h_stack)
        scores = torch.bmm(query, h_stack.transpose(1, 2))
        attn_weights = torch.softmax(scores, dim=-1)
        context = torch.bmm(attn_weights, h_stack)

        # Concatenate original hidden state with context vector
        combined = torch.cat((h_stack, context), dim=2)
        return self.fc(combined), None

# ==========================================
# TRAINING & EVALUATION SCRIPTS
# ==========================================

def train_model(model, dataset, epochs=50):
    # Determine the device the model is currently on
    device = next(model.parameters()).device

    dataloader = DataLoader(dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
    optimizer = optim.Adam(model.parameters(), lr=0.002)
    criterion = nn.CrossEntropyLoss(ignore_index=-1)

    model.train()
    for epoch in range(epochs):
        t_loss = 0
        for x, y in dataloader:
            # FIX: Move data to the same device as the model
            x, y = x.to(device), y.to(device)

            optimizer.zero_grad()
            out, _ = model(x)

            # CrossEntropyLoss expects (Batch, C, Seq) for 3D input
            loss = criterion(out.transpose(1, 2), y)
            loss.backward()
            optimizer.step()
            t_loss += loss.item()

        if (epoch+1) % 10 == 0:
            print(f"Epoch {epoch+1} Loss: {t_loss/len(dataloader):.4f}")
# --- Improved Weight Initialization Function ---
def init_weights(m):
    # For parameters in ParameterDict/ModuleList
    for name, param in m.named_parameters():
        if 'W' in name or 'weight' in name:
            # Xavier helps prevent the vanishing gradient in multi-layer RNNs
            nn.init.xavier_normal_(param.data if param.dim() > 1 else param.data.unsqueeze(0))
        elif 'b' in name or 'bias' in name:
            nn.init.constant_(param.data, 0)

# --- [Updated] Generate Names Function ---
def generate_names(model, dataset, num=50, temperature=0.85):
    model.eval()
    device = next(model.parameters()).device # Get model device
    generated_list = []

    for _ in range(num):
        name = "^"
        for i in range(25):
            # FIX: Move input_seq to device
            input_seq = torch.tensor([[dataset.char_to_int[c] for c in name]]).to(device)

            with torch.no_grad():
                logits, _ = model(input_seq)
                last_logit = logits[0, -1]
                probs = torch.softmax(last_logit / temperature, dim=0).cpu().numpy()

            idx = np.random.choice(len(probs), p=probs)
            char = dataset.int_to_char[idx]

            if char == "$": break
            if len(name) > 2 and char == name[-1] == name[-2]: continue
            name += char

        generated_list.append(name + "$")
    return generated_list

# def generate_names(model, dataset, num=50, temperature=0.8):
#     model.eval()
#     generated_list = []

#     for _ in range(num):
#         name = "^"
#         # Limit name length to 20 to prevent infinite loops
#         for i in range(20):
#             # Prepare input (last character)
#             x = torch.tensor([[dataset.char_to_int[name[-1]]]]).to(next(model.parameters()).device)

#             with torch.no_grad():
#                 logits, _ = model(x)
#                 # Take the last character's prediction
#                 last_logit = logits[0, -1]

#                 # Apply Temperature Scaling
#                 # formula: softmax(logits / T)
#                 probs = torch.softmax(last_logit / temperature, dim=0).cpu().numpy()

#             # Sample from the distribution instead of just taking the max (argmax)
#             idx = np.random.choice(len(probs), p=probs)
#             char = dataset.int_to_char[idx]

#             # Stop if we hit the end token
#             if char == "$":
#                 break

#             # Safety check: if the model repeats the same char 5 times, force stop
#             if len(name) > 5 and all(c == char for c in name[-4:]):
#                 break

#             name += char

#         generated_list.append(name + "$") # Ensure end token for consistency
#     return generated_list
# # def generate_names(model, dataset, num=50):
#     model.eval()
#     results = []
#     for _ in range(num):
#         name = "^"
#         for _ in range(20):
#             x = torch.tensor([[dataset.char_to_int[name[-1]]]])
#             out, _ = model(x)
#             probs = torch.softmax(out[0, -1], dim=0).detach().numpy()
#             idx = np.random.choice(len(probs), p=probs)
#             char = dataset.int_to_char[idx]
#             name += char
#             if char == "$": break
#         results.append(name)
#     return results


# # Execution
# dataset = NameDataset('trained_names.txt')
# configs = [
#     ("Vanilla RNN (2 Layers)", VanillaRNNScratch(dataset.vocab_size, 32, 64, 2)),
#     ("BLSTM (2 Layers)", BLSTMScratch(dataset.vocab_size, 32, 64, 2)),
#     ("Attention RNN (2 Layers)", AttentionRNNScratch(dataset.vocab_size, 32, 64))
# ]

# final_report = []
# for name, model in configs:
#     # 1. Apply initialization
#     model.apply(init_weights)

#     # 2. Train
#     print(f"\n--- Training {name} ---")
#     train_model(model, dataset, epochs=60) # Increased epochs for better convergence

#     # 3. Generate (Higher temperature for Attention/BLSTM)
#     temp = 1.0 if "Attention" in name else 0.85
#     gen = generate_names(model, dataset, num=50, temperature=temp)

#     # 4. Evaluate
#     unique = set(gen)
#     novel = [n for n in unique if n not in dataset.names]
#     novelty = (len(novel)/len(unique))*100 if unique else 0
#     diversity = len(unique)/len(gen)

#     final_report.append({
#         "Model": name,
#         "Params": sum(p.numel() for p in model.parameters()),
#         "Novelty %": f"{novelty:.2f}%",
#         "Diversity": f"{diversity:.2f}",
#         "Samples": [n.replace("^","").replace("$","").capitalize() for n in gen[:3]]
#     })

# print("\n", pd.DataFrame(final_report))


# 1. Define the search space
# You can add or remove values here to test different combinations
test_configs = [
    {"name": "Vanilla RNN", "class": VanillaRNNScratch, "hidden": 64, "lr": 0.002, "layers": 2, "temp": 0.8},
    {"name": "Vanilla RNN (High Temp)", "class": VanillaRNNScratch, "hidden": 64, "lr": 0.002, "layers": 2, "temp": 1.2},
    {"name": "BLSTM (Small)", "class": BLSTMScratch, "hidden": 32, "lr": 0.001, "layers": 2, "temp": 1.0},
    {"name": "Attention RNN", "class": AttentionRNNScratch, "hidden": 64, "lr": 0.002, "layers": 2, "temp": 1.0},
]

all_results = []

for cfg in test_configs:
    print(f"\n{'='*40}")
    print(f"RUNNING: {cfg['name']} | Hidden: {cfg['hidden']} | LR: {cfg['lr']}")
    print(f"{'='*40}")

    # Initialize Model
    if cfg['name'] == "Attention RNN":
        # Attention class in your code doesn't take num_layers as an init arg
        model = cfg['class'](dataset.vocab_size, 32, cfg['hidden'])
    else:
        model = cfg['class'](dataset.vocab_size, 32, cfg['hidden'], cfg['layers'])

    # Apply weights and move to device
    model.apply(init_weights)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Train (using a shorter epoch count for testing speed, adjust as needed)
    optimizer = optim.Adam(model.parameters(), lr=cfg['lr'])
    train_model(model, dataset, epochs=50)

    # Evaluate
    gen = generate_names(model, dataset, num=50, temperature=cfg['temp'])

    unique = set(gen)
    novel = [n for n in unique if n not in dataset.names]
    novelty = (len(novel)/len(unique))*100 if unique else 0
    diversity = len(unique)/len(gen)

    all_results.append({
        "Model": cfg['name'],
        "Hidden": cfg['hidden'],
        "LR": cfg['lr'],
        "Temp": cfg['temp'],
        "Novelty %": f"{novelty:.2f}%",
        "Diversity": f"{diversity:.2f}",
        "Sample": [n.replace("^","").replace("$","").capitalize() for n in gen[:2]]
    })

# Convert to DataFrame for easy reporting
comparison_df = pd.DataFrame(all_results)
print("\n--- Hyperparameter Comparison Report ---")
print(comparison_df)
```
